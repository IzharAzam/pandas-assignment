```python
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load all csv files using into a list of DataFrames using glob and list comprehension
frames_list = [pd.read_csv(csv) for csv in sorted(glob("states[0-9].csv"))]
```

```python
#concate the data as uc_census
us_census = pd.concat(frames_list)
```

```python
us_census.head(10)
```

```python
#drop unwanted column
us_census.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
us_census.head(10)
```

```python
#reset the index
us_census.reset_index(drop=True, inplace=True)
```

```python
us_census.head(10)
```

```python
us_census.dtypes
```

```python
#splitting GenderPop column into Men and Women Columns
genderpop = us_census.GenderPop.replace(['M','F'],'', regex=True).str.split('_')
men, women = genderpop.apply(lambda x: x[0]), genderpop.apply(lambda x: x[1])
us_census['Men'], us_census['Women'] = [men, women]
```

```python
us_census.head(10)
```

```python
us_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
us_census.dtypes
```

```python
#change newly created columns Men and Women data type to numeric
us_census.Men, us_census.Women = pd.to_numeric(us_census.Men), pd.to_numeric(us_census.Women)
us_census[['Men', 'Women']]
```

```python
# now dealing with NaN values in women population by filling it with values (TotalPop - Men)
us_census.Women.fillna(us_census.TotalPop - us_census.Men, inplace=True)
```

```python
us_census.head(10)
```

```python
us_census = us_census.replace('%','', regex=True)
```

```python
us_census.loc[:,'Hispanic':'Income'] = round((us_census.loc)[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
#plotting a scatter graph of Women and Income
plt.scatter(us_census.Women, us_census.Income, color=['red'])
plt.xlabel('Women',fontsize=18)
plt.ylabel('Income',fontsize=16)
plt.show()
```

```python
# check duplicated columns
us_census.duplicated() .value_counts()
```

```python
# drop duplicated rows and reset index to see the difference
us_census.drop_duplicates(inplace=True)
us_census.reset_index(drop=True, inplace=True)
us_census.duplicated() .value_counts()
```

```python
#plot again to see a perfect plot
plt.scatter(us_census.Women, us_census.Income, color=['green'])
plt.xlabel('Women')
plt.ylabel('Income')
plt.show()
```

```python
# races columns
us_census.iloc[:, 2:8].head()
```

```python
us_census.iloc[:, 2:8] = us_census.iloc[:, 2:8].replace('%','', regex=True).astype(float)
```

```python
#find nan values in races which are in only in Pacific race
us_census.Pacific.loc[us_census.Pacific.isnull() == True]
```

```python
#fill nan with mean of Pacific race
us_census.Pacific.fillna(us_census.Pacific.mean(), inplace=True)
us_census.Pacific.loc[us_census.Pacific.isnull() == True]
```

```python
# function to create Histograms of various races of US
def RaceHistogram(race, color):
    plt.hist(us_census[race],ec='black',color=color)
    plt.title(race+" Race")
    plt.ylabel("Frequency")
    plt.xlabel("Percentage of Population")
    plt.show()
```

```python
RaceHistogram('Hispanic', 'chocolate')
# White
RaceHistogram('White', 'whitesmoke')
# Black
RaceHistogram('Black', 'dimgrey')
# Native
RaceHistogram('Native', 'coral')
# Asian
RaceHistogram('Asian', 'saddlebrown')
# Pacific
RaceHistogram('Pacific', 'tan')
```
